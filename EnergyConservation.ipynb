```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
mass = 1 # 1 kg --> constanct
g = 9.8 # m/s^2 --> constant 

def Kinetic_Energy(mass, velocity_magnitude):
    """ Kineteic Energy is a function of 
        mass and magnitude of velocity 
    """
    y = 0.5*mass*velocity_magnitude**2
    return y

def Potential_energy( mass, g, vertical_displacement):
    """ Potential energy depends on 
        mass, gravitational acceleration, and vertical displacement
    """
    y = mass*g*vertical_displacement
    return y

Kinetic_Energy_0  = Kinetic_Energy(mass, 0) # initial velocity is 0
Potential_energy_0 = Potential_energy(mass, g, 1) # initial height is 1 meter
Total_Energy_0 = Kinetic_Energy_0 + Potential_energy_0 # Calculating total energy at the beginning 
array_of_vertical_displacement = np.linspace(1, 0, 100)
array_of_Potential_energy = Potential_energy(mass, g, array_of_vertical_displacement)
array_of_Kinetic_energy = Total_Energy_0 - array_of_Potential_energy; # total energy is conserved
array_of_total_energy = array_of_Kinetic_energy + array_of_Potential_energy # total energy is kinetic + potential
```

```python
plt.plot(array_of_vertical_displacement, array_of_Potential_energy, label ='Potential Energy')
plt.plot(array_of_vertical_displacement, array_of_Kinetic_energy, label='Kinetic Energy')
plt.plot(array_of_vertical_displacement, array_of_total_energy, label = 'Total Energy')
plt.xlabel('Vertical Displacement (meter)')
plt.ylabel('Energy (Joule)')
plt.legend()
plt.show()
```
